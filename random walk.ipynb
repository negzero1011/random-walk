```python
# ---------------------------- imports --------------------------------
import numpy as np
import matplotlib.pyplot as plt
# ---------------------------- imports --------------------------------

# ---------------------------- set seed for reproducibility -----------
np.random.seed(1)
# ---------------------------- set seed for reproducibility -----------

# clear the plot
plt.clf()

# ---------------------------- simulate 1000 random walks -------------
all_walks = []
for i in range(1000) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        
        # use one die
        dice = np.random.randint(1, 7)
        if dice <= 3:
            step = step - 1
        else:
            step = step + 1


        random_walk.append(step)
    all_walks.append(random_walk)
# ---------------------------- simulate 1000 random walks -------------

# ---------------------------- visualize results ----------------------
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
# ---------------------------- visualize results ----------------------
```

```python
# ---------------------------- imports --------------------------------
import numpy as np
import matplotlib.pyplot as plt
# ---------------------------- imports --------------------------------

# ---------------------------- set seed for reproducibility -----------
np.random.seed(1)
# ---------------------------- set seed for reproducibility -----------

# clear the plot
plt.clf()

# ---------------------------- simulate 1000 random walks -------------
all_walks = []
for i in range(1000) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        
        # use two dice
        dice = np.random.randint(2, 13)
        if dice <= 6:
            step = step - 1
        elif dice == 7:
            step = step
        else:
            step = step + 1


        random_walk.append(step)
    all_walks.append(random_walk)
# ---------------------------- simulate 1000 random walks -------------

# ---------------------------- visualize results ----------------------
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
# ---------------------------- visualize results ----------------------
```
